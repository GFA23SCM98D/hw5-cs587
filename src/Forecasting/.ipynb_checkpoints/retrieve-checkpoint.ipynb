```python
import requests
import datetime
import json
```

```python
# GitHub API的基本URL
base_url = "https://api.github.com/repos/"

# 仓库列表
repos = [
    "openai/openai-cookbook",
    "openai/openai-python",
    "openai/openai-quickstart-python",
    "milvus-io/pymilvus",
    "SeleniumHQ/selenium",
    "golang/go",
    "google/go-github",
    "angular/material",
    "angular/angular-cli",
    "SebastianM/angular-google-maps",
    "d3/d3",
    "facebook/react",
    "tensorflow/tensorflow",
    "keras-team/keras",
    "pallets/flask"
]
```

```python
# GitHub Token
token = "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"

# 获取过去两个月的起始日期和结束日期
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=60)

# 存储所有信息的字典
data_dict = {}
```

```python
# 遍历仓库列表
for repo in repos:
    # 构建API请求URL
    url = base_url + repo

    # 发送GET请求获取仓库信息
    headers = {
        "Authorization": "token " + token
    }
    response = requests.get(url, headers=headers)
    data = response.json()

    # 获取stars、forks、issues等信息
    stars = data["stargazers_count"]
    forks = data["forks_count"]
    issues = data["open_issues_count"]

    # 构建API请求URL，获取过去两个月的issues
    issues_url = url + "/issues"
    params = {
        "state": "all",
        "since": start_date.isoformat()
    }

    # 发送GET请求获取issues信息
    issues_response = requests.get(issues_url, params=params, headers=headers)
    issues_data = issues_response.json()

    # 统计issues created和issues closed数量
    issues_created = 0
    issues_closed = 0

    for issue in issues_data:
        created_at = datetime.datetime.strptime(issue["created_at"], "%Y-%m-%dT%H:%M:%SZ")
        if created_at >= start_date:
            issues_created += 1
        if issue["state"] == "closed":
            closed_at = datetime.datetime.strptime(issue["closed_at"], "%Y-%m-%dT%H:%M:%SZ")
            if closed_at >= start_date:
                issues_closed += 1

    # 构建API请求URL，获取过去两个月的pulls
    pulls_url = url + "/pulls"
    params = {
        "state": "all",
        "since": start_date.isoformat()
    }

    # 发送GET请求获取pulls信息
    pulls_response = requests.get(pulls_url, params=params, headers=headers)
    pulls_data = pulls_response.json()

    # 统计pulls数量
    pulls = len(pulls_data)

    # 构建API请求URL，获取过去两个月的commits
    commits_url = url + "/commits"
    params = {
        "since": start_date.isoformat()
    }

    # 发送GET请求获取commits信息
    commits_response = requests.get(commits_url, params=params, headers=headers)
    commits_data = commits_response.json()

    # 统计commits数量
    commits = len(commits_data)

    # 构建API请求URL，获取过去两个月的branches
    branches_url = url + "/branches"

    # 发送GET请求获取branches信息
    branches_response = requests.get(branches_url, headers=headers)
    branches_data = branches_response.json()

    # 统计branches数量
    branches = len(branches_data)

    # 构建API请求URL，获取过去两个月的contributions
    contributors_url = url + "/contributors"

    # 发送GET请求获取contributors信息
    contributors_response = requests.get(contributors_url, headers=headers)
    contributors_data = contributors_response.json()

    # 统计contributions数量
    contributors = len(contributors_data)

    # 构建API请求URL，获取过去两个月的releases
    releases_url = url + "/releases"
    params = {
        "since": start_date.isoformat()
    }

    # 发送GET请求获取releases信息
    releases_response = requests.get(releases_url, params=params, headers=headers)
    releases_data = releases_response.json()

    # 统计releases数量
    releases = len(releases_data)

    # 将信息存储到字典中
    data_dict[repo] = {
        "stars": stars,
        "forks": forks,
        "issues": issues,
        "issues_created": issues_created,
        "issues_closed": issues_closed,
        "pulls": pulls,
        "commits": commits,
        "branches": branches,
        "contributors": contributors,
        "releases": releases
    }
```

```python
# 将字典保存为JSON文件
with open("github_info.json", "w") as file:
    json.dump(data_dict, file, indent=4)
```
